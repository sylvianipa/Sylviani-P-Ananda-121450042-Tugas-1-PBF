###- **DATASET 2 bike-sharing**

1. Data Loading

```python
import pandas as pd #berfungsi untuk mengimport pandas untuk mempermudah penulisan code

df_bike = pd.read_csv(("/content/bike-sharing.csv"), index_col="instant") #berfungsi untuk membaca dan memuat file CSV ke dalam data frame dan menjadikan sebagai indeks dataframe
df_bike.head(3) #berfungsi untuk menampilkan 3 baris pertama dari data frame
```

2. Visualization

```python
#Visualisation
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(1, figsize=(10, 5))
sns.scatterplot(
    data=df_bike,
    x="temp",
    y="hum",
    hue="season",
    size="windspeed",
    palette="Set2"
)
```

3. Normalizing

```python
from sklearn import preprocessing

# Pilih kolom-kolom yang akan digunakan untuk normalisasi
columns_for_normalization = ["temp", "atemp", "hum", "windspeed", "casual", "registered", "cnt"]

# Pisahkan kolom-kolom yang akan dinormalisasi
X = df_bike[columns_for_normalization]

# Normalisasi data menggunakan Min-Max Scaling
scaler = preprocessing.MinMaxScaler()
X_norm = scaler.fit_transform(X)

# Tampilkan contoh hasil normalisasi
print("hasil normalisasi:")
print(pd.DataFrame(X_norm, columns=columns_for_normalization).head())
```

4. Elbow Method

```python
import numpy as np #berfungsi untuk mengimpor numpy untuk operasi numerik
from sklearn.cluster import KMeans #berfungsi untuk mengimpor modul KMeans untuk melakukan klasterisasi dengan algortima KMeans

def elbow_plot(data,clusters): #berfungsi untuk mendefinisikan fungsi elbow dengan 2 paramter dan clusters
    inertia = [] #berfungsi untuk membuat list kosong untuk menyimpan nilai inertia
    for n in range(1, clusters):
        algorithm = KMeans(
            n_clusters=n,
            init="k-means++",
            random_state=125, #berfungsi untuk membuat objek algorithm dengan model KMeans menggunakan inisialisasi k-means++ dan seed random state 125
        )
        algorithm.fit(data) #berfungsi untuk melatih model KMeans pada data
        inertia.append(algorithm.inertia_) #berfungsi untuk menambahkan nilai inertia dari model ke dalam list inertia
    # plot
    plt.plot(np.arange(1 , clusters) , inertia , 'o') #berfungsi untuk membuat plot nilai inertia menggunakan simbol o
    plt.plot(np.arange(1 , clusters) , inertia , '-' , alpha = 0.5) #berfungsi untuk membuat plot garis lurus yang menghubungkan nilai inertia
    plt.xlabel('Number of Clusters') , plt.ylabel('Inertia') #berfungsi untuk memberikan sumbu x sebagai number of cluster dan sumbu y sebagai inertia
    plt.show(); #berfungsi untuk menampilkan plot elbow
```

```python
elbow_plot(X_norm,10) #berfungsi untuk mengklasterisasi jumlah klaster yang sudah dinormalisasikan dari 1 hingga 9
```

5. K-Means Clustering

```python
algorithm = KMeans(n_clusters=3, init="k-means++", random_state=125) #berfungsi untuk membuat objek KMeans menggunakan inisialisasi dan seed random sebesar 125
algorithm.fit(X_norm)  #berfungsi untuk melatih model yang sudah dinormalisasikan dengan mengelompokkan data ke dalam kluster
labels = algorithm.labels_  #berfungsi untuk mendapatkan label kluster untuk setiap sampel data KMeans yang telah dilatih
```

```python
sns.scatterplot(data = X, x = 'temp', y='hum', hue = labels, palette="Set2");
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.scatterplot(
    data=df_spotify,
    x="danceability",
    y="energy",
    hue="mode",
    size="valence",
    palette="Set2"
)
```

```python
from sklearn import preprocessing

# Drop non-numeric and irrelevant columns
columns_to_drop = ["mode", "liveness", "track_id", "track_name", "track_artist", "track_album_id", "track_album_name", "track_album_release_date", "playlist_name", "playlist_id", "playlist_genre", "playlist_subgenre"]
X = df_spotify.drop(columns_to_drop, axis=1)

# Handle missing values (replace NaN with mean, median, or any other strategy)
X.fillna(X.mean(), inplace=True)  # You can choose a different strategy if needed

# Normalization using Min-Max Scaling
X_norm = preprocessing.normalize(X)
```

```python
#4. Elbow Method
import numpy as np
from sklearn.cluster import KMeans

def elbow_plot(data,clusters):
    inertia = []
    for n in range(1, clusters):
        algorithm = KMeans(
            n_clusters=n,
            init="k-means++",
            random_state=125,
        )
        algorithm.fit(data)
        inertia.append(algorithm.inertia_)
    #Plot
    plt.plot(np.arange(1 , clusters), inertia, 'o')
    plt.plot(np.arange(1 , clusters), inertia, '-', alpha = 0.5)
    plt.xlabel('Number of Clusters'), plt.ylabel('Inertia')
    plt.show();
```

```python
elbow_plot(X_norm, 10)
```

```python
algorithm = KMeans(n_clusters=3, init="k-means++", random_state=125)
algorithm.fit(X_norm)
labels = algorithm.labels_
```
